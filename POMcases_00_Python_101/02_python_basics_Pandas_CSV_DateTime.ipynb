## Handling Missing Values <a class="anchor" id="handlingmissingvalues"></a>

In Python, particularly with pandas and NumPy, handling missing values is facilitated by the usage of `NaN` (Not a Number) and `None` to represent such values. The pandas library provides various functions to manage, identify, and impute these missing values in a DataFrame.

- Identifying and Representing Missing Values
- Detecting Missing Values
- Managing Missing Values


````python
import pandas as pd
import numpy as np
````

````python
# A missing value is typically represented as NaN in Python, within a NumPy array or a pandas DataFrame.
missing_value = np.nan
type(missing_value)
````

Arrays (or lists) and DataFrames in Python can contain `NaN` values, which might necessitate handling or imputing these missing values before performing certain operations.


````python
# Example array with missing values
x = [1, 2, np.nan, 3]

# Creating a DataFrame with missing values
df1 = pd.DataFrame({'a': [1, 2, np.nan], 'b': ['a', 'b', None]})
````

To check for missing values in Python, `isna()` or `isnull()` can be used for a DataFrame, and `np.isnan()` for a NumPy array. 

- - - 

# Checking for missing values <a class="anchor" id="checkingmissingvalues"></a>
pd.isna(1), pd.isna(np.nan), pd.isna(x), df1.isna()

To ignore or skip missing values in a list or array, you might filter them out explicitly.

````python
# Skipping missing values
[i for i in [1, np.nan, 2, np.nan] if not pd.isna(i)]
````

You can replace `NaN` or `None` values using `replace()` in pandas, which returns a new object with the missing values replaced by a specified value.

````python
# Replacing missing values
df1.replace(np.nan, value=-1)
````

When dealing with DataFrames, you can modify them in place using `inplace=True` in certain functions like `replace()`, which avoids the need to assign the result back to a variable.


````python
# Replacing values in-place in a DataFrame column
df1['a'].replace(np.nan, value=100, inplace=True)
````

To remove rows containing missing values from a DataFrame, `dropna()` can be utilized. The `inplace` parameter determines whether the operation is performed in-place or a new DataFrame is returned.

````python
# Removing rows with missing values
y = df1.dropna()
df1.dropna(inplace=True)
````

## Joining DataFrames <a class="anchor" id="joindata"></a>



### Rules for the `on` keyword argument:

* A single string if joining on one column with the same name, e.g., `on='id'`.
* A list of strings if joining on multiple columns with the same name, e.g., `on=['id1', 'id2']`.
* For joining on columns with different names in the two dataframes, use the `left_on` and `right_on` keyword arguments. For example, to join on dataframe `df1`'s column `'a'` and dataframe `df2`'s column `'b'`, use `left_on='a'` and `right_on='b'`.

```python
import pandas as pd

# Example usage:
df_merged = pd.merge(left=df1, right=df2, on='id')
df_merged = pd.merge(left=df1, right=df2, on=['id1', 'id2'])
df_merged = pd.merge(left=df1, right=df2, left_on='a', right_on='b')


````python
# inner join by default, same values are joined but also missing is joined

merged_inner = pd.merge(left=df1, right=df2, how='inner', on='ID')

merged_inner
````

````python
merged_left = pd.merge(left=df1, right=df2, how='left', on='ID')

merged_left
````

````python
merged_right = pd.merge(left=df1, right=df2, how='right', on='ID')

merged_right
````

````python
merged_outer = pd.merge(left=df1, right=df2, how='outer', on='ID')

merged_outer
````

# DateTime <a class="anchor" id="datetime"></a>


We would like to analyze the sales regarding the times when sales occurred during the day. First, we need to extract the times from the timestamp.

In Python, using pandas, we can achieve this by accessing the datetime properties directly once the column is in the datetime type. We'll read our sales data, extract relevant time information, and perform some basic analysis on it, such as calculating the sum, mean, and standard deviation of sales at different times.

Additionally, we will create new columns in the DataFrame to hold the extracted hour and day information from the timestamp, which allows for further detailed analysis.


````python
import pandas as pd

# Reading the CSV file with specified header and types
sales = pd.read_csv("data/BerlinSales.csv", 
                    names=["DateTime", "Sales"],
                    dtype={"DateTime": str, "Sales": int},
                    parse_dates=["DateTime"],
                    date_parser=lambda x: pd.to_datetime(x, format="%d.%m.%Y %H:%M"))

# Extracting the hour and day from the DateTime column and creating new columns for them
sales['Hour'] = sales['DateTime'].dt.hour
sales['Day'] = sales['DateTime'].dt.day

# Displaying the DataFrame
sales.head()
````

With our newly created columns, we can now perform some basic analysis, like calculating the sum, mean, and standard deviation (`std`) of sales at different times of the day.

We will use the `groupby` method for this, allowing us to group our data by the hour of sale, and calculate various statistics for these groups. Additionally, we can also differentiate our analysis for different days.


````python
# Calculating and displaying the sum of Sales grouped by Hour
sales_sum_by_hour = sales.groupby('Hour')['Sales'].sum()
print("Sum of Sales by Hour:")
print(sales_sum_by_hour)
````

````python
# Calculating and displaying the mean of Sales grouped by Hour
sales_mean_by_hour = sales.groupby('Hour')['Sales'].mean()
print("\nMean of Sales by Hour:")
print(sales_mean_by_hour)
````

````python
# Calculating and displaying the standard deviation of Sales grouped by Hour
sales_std_by_hour = sales.groupby('Hour')['Sales'].std()
print("\nStandard Deviation of Sales by Hour:")
print(sales_std_by_hour)
````

````python
# Calculating and displaying the sum of Sales grouped by Day and Hour
sales_sum_by_day_hour = sales.groupby(['Day', 'Hour'])['Sales'].sum()
print("\nSum of Sales by Day and Hour:")
print(sales_sum_by_day_hour)
````
