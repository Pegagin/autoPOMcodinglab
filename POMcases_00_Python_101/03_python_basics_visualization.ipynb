If the label clashes with our graph, the label can be moved by changing the scaling of the y-axis...

```python
plt.figure(figsize=(7,5))
plt.plot(x2, y2)
plt.title("Line graph with two lines")
plt.legend(["line 1", "line 2"])
plt.ylim([0, 1.5])  # set the scale of the y-axis
plt.show()
```

When you're placing a legend outside the plot, it is typically recommended to use both `bbox_to_anchor` and `loc` together.

The `bbox_to_anchor` argument specifies the box that is used to position the legend in conjunction with `loc`.
`bbox_to_anchor` moves the legend box around according to the coordinates you specify, and loc determines which part of the bounding box of the legend should be placed at the provided coordinates.

```python
plt.figure(figsize=(7,5))
plt.plot(x2, y2)
plt.title("Line graph with two lines")
plt.legend(["line 1", "line 2"], bbox_to_anchor=(1.2, 1))
plt.show()
```

Add a label for the x- and the  y-axis.

```python
plt.plot(x2, y2)
plt.title("Line graph with two lines")
plt.xlabel("x-axis")  # label for the x-axis
plt.ylabel("y-axis")  # label for the y-axis
plt.legend(["line 1", "line 2"])
plt.show()
```

You can customize many aspects of your plot in Matplotlib. For example:

`Change line color:` You can specify the color of each line in the plot command using the color parameter.

`Change line width:` The width of each line can be changed using the linewidth parameter.

`Change background color:` The background color of the plot can be set using the set_facecolor method on the current axes, obtained with plt.gca().

Find an overview of named colors here: https://matplotlib.org/3.1.0/gallery/color/named_colors.html

```python
plt.plot(x2, y2[:,0], color='darkslategrey', linewidth=0.5)  # Change color and width of line 1
plt.plot(x2, y2[:,1], color='red', linewidth=0.5)  # Change color and width of line 2
plt.title("Line graph with two lines")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["line 1", "line 2"])
plt.gca().set_facecolor('floralwhite')  # Change background color
plt.show()
```

## 2) Draw scatter plots<a class="anchor" id="scatter"></a>

Plot a basic scatter plot.

```python
plt.figure(figsize=(7,5))
plt.scatter(x2, y2[:,0], color='steelblue', label='data 1')
plt.scatter(x2, y2[:,1], color='firebrick', label='data 2')
plt.legend()
plt.show()
```

In order to personalize your graph, various sets of attributes can be adjusted, enhancing the visual appeal and readability of your plots when using the Python plotting library `matplotlib`. These attributes are broadly categorized as follows:

* **Series Attributes:** Related to the content of your graph, such as `color`, `linestyle`, `marker`, or `markersize`.
* **Plot Attributes:** Related to the general layout of your graph, like `figsize`, `title`, or `legend`.
* **Axes Attributes:** Pertaining to the layout of your axes, and involve properties like `xlabel`, `ylabel`, `xlim`, `ylim`, or `xticks`.
* **Subplot Attributes:** These attributes, like `sharex`, `sharey`, or `subplot_kw`, manage subplot layouts when working with multiple plots in a single figure.

In addition to the aforementioned attributes, when constructing scatter charts with `matplotlib`, consider the following customizations:

* **Markers:** Alter the shape of markers using the `marker` argument and adjust their size with the `s` argument.
* **Edge color:** Modify the edge color of markers via the `edgecolor` argument.
* **Transparency:** Adjust marker transparency using the `alpha` argument.
* **Title, labels, and legend:** Add a title with `plt.title()`, axis labels with `plt.xlabel()` and `plt.ylabel()`, and a legend with `plt.legend()`.
* **Grid and limits:** Incorporate a grid with `plt.grid()`, and establish axis limits using `plt.xlim()` and `plt.ylim()`.

Below is a Python code snippet demonstrating how these attributes can be utilized with `matplotlib`. Feel free to experiment by commenting/uncommenting different lines to observe their effects on the plot.


```python
# Individualized scatter plot
plt.figure(figsize=(7,5))

# Scatter plots with customized markers, edge colors, and transparency
plt.scatter(x2, y2[:,0], color='steelblue', edgecolor='black', alpha=0.7, s=100, marker='o', label='data 1')
plt.scatter(x2, y2[:,1], color='firebrick', edgecolor='black', alpha=0.7, s=100, marker='s', label='data 2')

# Title and labels
plt.title("Individualized Scatter Plot", fontsize=16)
plt.xlabel("X Values", fontsize=14)
plt.ylabel("Y Values", fontsize=14)

# Legend
plt.legend(fontsize=12)

# Grid and limits
plt.grid() # Add gridlines
plt.xlim([0, 11])
plt.ylim([0, 1.5])

plt.show()
```

## 3) Draw bar charts<a class="anchor" id="bar"></a>

Draw a simple bar chart, individualize the ticks of the x-axis.

#### Static approach

`plt.xticks([0,1,2],["Data 1","Data 2","Data 3"])` sets the ticks at positions 0, 1, and 2 on the x-axis to be labelled as "Data 1", "Data 2", and "Data 3" respectively. 

#### Dynamic approach (advanced)

`labels=[f'data {i+1}' for i in x3]` is generating a list of strings, where each string is of the form `data x`, with `x` being each element in the list x3 incremented by 1.

The expression `f'data {i+1}` is a formatted string literal, often referred to as an f-string. The `{i+1}` inside the f-string is a placeholder that gets replaced by the value of `i+1`.

```python
import numpy as np

y3 = np.random.rand(3) # generates 3 random values
x3 = np.arange(len(y3)) # x = [0 1 2]

plt.figure(figsize=(7, 5))
plt.bar(x3, y3, color='deepskyblue')
plt.xticks([0,1,2],["Data 1","Data 2","Data 3"])  # static approach
#plt.xticks(x3, labels=[f'data {i+1}' for i in x3])  # dynamic approach
plt.title('My bar chart')
plt.xlabel('Data points')
plt.ylabel('Random value')
plt.show()
```

Draw a grouped bar chart. This involves a bit of manual adjustment of the bar positions.


In the for loop, each iteration plots a group of bars, shifting them along the x-axis to prevent overlap, and assigns a label for the legend, thereby creating the grouped bar chart effect by positioning adjacent bars representing different data groups.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example data and set width for bars
y3 = np.random.rand(4, 3)
bar_width = 0.2
x3 = np.arange(y3.shape[0])

# Initialize plot and create bars in a loop
plt.figure(figsize=(7, 5))
for i in range(y3.shape[1]):
    plt.bar(x3 + i * bar_width, y3[:, i], width=bar_width, label=f'group {i+1}')

# Label and show the plot
plt.title('My grouped bar chart')
plt.xlabel('Rows')
plt.ylabel('Random value')
plt.xticks(x3 + bar_width, [f'Row {i+1}' for i in range(y3.shape[0])])  # dynamic approach
plt.legend()
plt.show()
```

Draw a stacked bar chart.

```python
# in matplotlib, stacked bar charts are done by plotting the bars on top of each other
plt.figure(figsize=(7, 5))

for i in range(y3.shape[1]):
    plt.bar(x3, y3[:, i], bottom=np.sum(y3[:, :i], axis=1), label=f'group {i+1}')

plt.title('My stacked bar chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

## 4) Draw violin plots<a class="anchor" id="statistical"></a>

Draw violin plots to indicate the distribution of data, see some explanation on how to read the plot here: https://mode.com/blog/violin-plot-examples.

```python
y4 = np.random.rand(100,4) # create 4x100 random values

fig, ax = plt.subplots()
ax.violinplot(y4) # add mean or median with 'showmeans=True' or 'showmedians=True' inside '.violinplot(...)'
ax.set_title('Violin plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.xticks([1,2,3,4],["Series 1", "Series 2", "Series 3", "Series 4"]) # set the x-tick labels on the plot

plt.show()
```

Add a boxplot of the same data (on top), see some explanation on how to interpret boxplots here: https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51

```python
fig, ax = plt.subplots()
ax.violinplot(y4) # remove this line to see the boxplots only
ax.boxplot(y4)
ax.set_title('Violin- and Boxplot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.xticks([1,2,3,4],["Series 1", "Series 2", "Series 3", "Series 4"])
plt.show()
```

## 5) Combine different plots<a class="anchor" id="ensemble"></a>

To display multiple diagrams on one figure in Matplotlib, you can use the `concept of subplots`. With subplots you can divide a figure into a grid structure and fill each grid field with its own diagram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x5 = np.arange(1, 11)
y5 = np.arange(1, 11)

# Initialize a 2x2 subplot figure
fig, axs = plt.subplots(2, 2, figsize=(7, 5)) # use 'axs' when creating a figure with multiple subplots

# Line graph
axs[0, 0].plot(x5, y5)
axs[0, 0].set_title('My line graph')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Scatter plot
axs[0, 1].scatter(x5, y5, color='deepskyblue')
axs[0, 1].set_title('My scatter chart')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')

# Bar chart
axs[1, 0].bar(x5, y5, color='deepskyblue')
axs[1, 0].set_title('My bar chart')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')

# Stacked bar chart
for i in range(10):
    axs[1, 1].bar(x5, y5, bottom=i*y5) # 'bottom' determines the 'starting point' or the 'base' of each bar
axs[1, 1].set_title('My stacked bar chart')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')

plt.tight_layout()  # this will improve the spacing between the plots
plt.show()
```

## 6) Plot DataFrames with Pandas<a class="anchor" id="dataframes"></a>

```python
import pandas as pd # for DataFrames
```

Pandas is designed to work with data that is organized in tables, and it excels at performing operations on tabular data.

The good news is that `Pandas dataframes` have built-in plotting methods that leverage `Matplotlib` under the hood. This makes it easy to visualize data directly from a dataframe.

Since we are still working with `Matplotlib`, commands like `plt.xticks, plt.title, ...` are still valid.

Here are some examples with the given dataframe:

Import data and create a DataFrame.

```python
df = pd.read_csv("data/jan_sales.csv") # create DataFrame

df.head()
```

```python
# Line plot of Units Sold
df.plot(kind='line', x='Date', y='Units Sold')
plt.title("Units Sold Over Time")
plt.show()
```

```python
# Bar plot of Total Sales by Product
df.plot(kind='bar', x='Product', y='Total Sales')
plt.title("Total Sales by Product")
plt.show()
```

```python
# Histogram of 'Unit Price'
df['Unit Price'].plot(kind='hist', edgecolor='black')
plt.title("Histogram of Unit Price")
plt.xlabel("Unit Price")
plt.show()
```

```python
# Scatter plot of Unit Price vs Units Sold
df.plot(kind='scatter', x='Unit Price', y='Units Sold')
plt.title("Scatter Plot: Unit Price vs Units Sold")
plt.xlabel("Unit Price")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()
```
