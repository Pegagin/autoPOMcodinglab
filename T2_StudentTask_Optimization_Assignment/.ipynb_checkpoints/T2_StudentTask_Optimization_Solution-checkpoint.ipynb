## Part 1: Optimize using Cbc package 

### Data

```python
# Import .csv file, sort by date, and save into DataFrame 
df_sort_by_task = CSV.read("data/WorkTasks_csv.csv", DataFrame, dateformat=dateformat"m/d/yyyy");
df_sort_by_date = sort(df_sort_by_task, "Due date")   # sort depending on date
first(df_sort_by_date,4)
```

```python
# Data input for mathematical program
d_all = collect(Date(2020,5,11):Day(1):Date(2020,5,31)); # Create a range of dates starting from 11.5.20 to 31.5.20
T_all = length(d_all); # Number of days in this range

T_plan = 7; # define length of plan

# DataFrame of tasks for optimization; 1st day is May 11
df_optimize=subset(df_sort_by_task, :"Due date" => ByRow(>=(Date(2020,5,11))); skipmissing=true)
df_optimize=subset(df_optimize, :"Due date" => ByRow(<=(Date(2020,5,11+T_plan-1))))

types = unique(df_optimize[:,3]); # take the 3rd column of df_optimize and return a array of all types there are
types = sort(types); # sort depending on product type
num_types = length(types); # calculate number of product types
num_tasks = size(df_optimize)[1]; # calculate number of tasks (= number of elements in 1st column of DataFrame)

cap = 10; # initialize capacity

d_task = df_optimize[:,2]; # extract 2nd column of optimize
n_task = df_optimize[:,1]; # same with 1st column

first(df_optimize,20)
```

```python
# what is happening in this cell?
task_types= Array{Bool}(undef, (num_tasks, num_types)) # create 2D (67x5) Array with random values

for i in 1:num_types
    task_types[:,i] = (df_optimize[:, 3] .== types[i]); # Take the i-rd column and check whether the Type matches with the df_opti table
end
```

## Model

>We formulate a MILP (Mixed-Integer Linear Programming) to minimize lateness (positive and negative violations) for the defined lenght of the plan of one week.

```python
MILP = Model(Cbc.Optimizer); # initialize mathematical program with name MILP; ich würd's etwas feingranularer machen.
#set_optimizer_attribute(MILP, "logLevel", 0); # no solver output
```

```python
@variable(MILP, X[1:T_plan, 1:num_tasks], Bin); # initialize binary assignment variables (task --> period)
@variable(MILP, DueDev[1:num_tasks] >= 0); # initialize lateness variable
```

```python
@objective(MILP, Min, sum(DueDev[i] for i in 1:num_tasks)); # objective function: minimize total lateness
```

```python
# what is happening here?
@constraint(MILP, Earliness[i=1:num_tasks], DueDev[i]>= sum((Dates.value(d_task[i]-d_all[t])*X[t,i]) for t in 1:T_plan) );
@constraint(MILP, Tardiness[i=1:num_tasks], DueDev[i]>= sum((Dates.value(d_all[t]-d_task[i])*X[t,i]) for t in 1:T_plan) );

# Capacity constraint: 
@constraint(MILP, CapacityRestriction[i = 1:T_plan, 1:num_tasks], X[i]>=)
# add a new set of constraints called CapacityRestriction: in each period there should not be more assigned tasks than there is capacity 

# Schedule each task
# add a new constraint called TaskExactlyOnce: each task has to be assigend to exactly one period

# what is happening here?
@constraint(MILP, OnlyOneHard[t=1:T_plan], sum( X[t,i]*task_types[i,1] for i in 1:num_tasks) <= 1);
```

```python
@time optimize!(MILP);
```

## Results

```python
println("Objective value: ", JuMP.objective_value(MILP), " found after ", round(JuMP.solve_time(MILP), digits=2), " seconds. Relative gap is: ", JuMP.relative_gap(MILP))
```

```python
#Set number of periods for the analysis
T_analyze=7;
```

```python
# print lateness per order
JuMP.value.(DueDev)[1:T_analyze]
```

```python
# number of jobs in daily bins
sum(JuMP.value.(X),dims=2)[1:T_analyze] #dims=2 makes sure that sum is computed over rows of array (=periods)
```

```python
# Write assignment of tasks to periods to an array, print result

X_all=JuMP.value.(X); # copy results to matrix to speed up the runtime

X_results= Array{Union{Nothing, Int}}(nothing, (T_analyze, cap)); #Initialize empty array with dimension (T_plan x cap)
for period in 1:T_analyze # for each period
    print("day ",period," : ")
    i=1;
    for task in 1:num_tasks # for each task
        if X_all[period, task] > 0 #check, if task is assigned to period
            X_results[period, i] = n_task[task]; # write number of assigend tasks into X_results array of period
            print(n_task[task]," ")
            i+=1;
        end
    end
    println()
end       
```

```python
# print array of plan
X_results[1:T_analyze, :]
```

### Plots

```python
# plot the scheduled tasks for the first week

X_all=JuMP.value.(X); # copy results to matrix

plot = groupedbar(X_all[1:T_analyze,:], legend=false, bar_position = :stack, yticks=1:cap)
for xpos in 1:T_analyze
    for ypos in 1:cap
        annotate!(xpos+0.1, ypos-0.5, text(X_results[xpos, ypos], :black, :right, 8))
    end
end
plot
```

```python
# now add a plot where the colors match the type of the item

plotten = Matrix([(0) for i in 1:T_analyze, j in 1:cap]); # initialize array of zeros
farbgebung = Matrix([(:white) for i in 1:1, j in 1:T_analyze*cap]); #what is happening here?

farben = [:red :aquamarine3 :teal :deepskyblue4 :midnightblue]
for t in 1:T_analyze # all periods
    j = t * cap; # last position in plan of current period
    for i in 1:cap # loop over positions in plan
        for s in 1:num_types # loop over types
            if X_results[t,i]!=nothing # order assigned to position
                if df_sort_by_task[X_results[t,i],3] == types[s] # compare type of task in current position of plan with type s
                    plotten[t,i] = 1.0; # indicate that position i in period t is filled by a task
                    farbgebung[j] = farben[s]; # save colour depending on the type of task #j 
                end
            end
        end
        j = j-1; 
    end
end

# initialize plot for all positions filled with tasks in white colour
plot = groupedbar(plotten[1:T_analyze,:], legend=false, bar_position = :stack, yticks=1:cap, colour = :white)

# add colours
for n in 1:T_analyze # for each period
    first_job=10 * (n-1); # point to the first job in current period 
    plot = groupedbar!(vcat(vcat(Matrix([(0) for i in 1:(n-1), j in 1:cap]),plotten[1:1,:]),Matrix([(0) for i in n+1:T_analyze, j in 1:cap])), #add plots for each period; inner logic: merge three arrays of dimension (t_plan x cap): 1st with zeros, 2nd (active period n) with ones for each filled position in plan; 3rd with zeros
        legend=false, bar_position = :stack, yticks=1:cap, 
        colour = [farbgebung[first_job + 1] farbgebung[first_job + 2] farbgebung[first_job + 3] farbgebung[first_job + 4] farbgebung[first_job + 5] farbgebung[first_job + 6] farbgebung[first_job + 7] farbgebung[first_job + 8] farbgebung[first_job + 9] farbgebung[first_job + 10]]) # add colour information depending on typ of job
end

# add taks numbers
for xpos in 1:T_analyze
    for ypos in 1:cap
        annotate!(xpos+0.1, ypos-0.5, text(X_results[xpos, ypos], :white, :right, 8))
    end
end
plot
```

### Production planning <a class="anchor" id="pp"></a>


#### Sets and indices

$j \in (1,..,J) \hspace{20mm}$ Tasks

#### Decision variables

$X_{j,t}  \in (0,1) \hspace{20mm}$ Assignment of task $j$ to period $t$

$abc   \geq0 \hspace{20mm}$ xyz

#### Parameters 

$type_j \hspace{39mm}$   Type of taks $j$ 

#### Objective function:
$ \hspace{30mm} \max \quad abc = \displaystyle\sum_{i=1}^I(p_i - c_i^v)\cdot m_i $

#### Subject to:
1) **xyz:** $\hspace{29mm} \sum_{i=1}^I(r_{ij}\cdot X_i) \leq cap_j \hspace{29mm} \forall j \in J $

## Part 2: Optimize using Gurobi solver

When solving the model for 20 periods, you saw the solution time to increase significantly. In generel, larger instances of your problem or more complex models can lead to long solving times. One way to improve solving efficiency is to use a better solver than Cbc. A very good commercial solver is *Gurobi*. 

You can download Gurobi [here](https://www.gurobi.com/downloads/gurobi-optimizer-eula/).

As Gurobi is a commercial solver, you need to get a license to run it. You can request a free academic license [here](https://www.gurobi.com/downloads/end-user-license-agreement-academic/)

Follow the instructions on the website to install the license on your computer. Make sure Gurobi is installed before the license. 

This license is usually valid for two months, but a new one can be requested when the old one expires. 

Just like the Cbc solver needs the Cbc package to be run in Julia, the Gurobi solver needs a Gurobi package. Add the package and use it. 

```python
Pkg.add("Gurobi"); # Needed for initial use of Gurobi
using Gurobi;
```

Now you can solve the model with Gurobi, by changing the optimizer when creating the model. 
Try and see how the solving time for 20 periods changes. 

```python
MILP = Model(Gurobi.Optimizer); # Here Cbc was replaced by Gurobi

@variable(MILP, X[1:T_plan, 1:num_tasks], Bin); 
@variable(MILP, DueDev[1:num_tasks] >= 0);
@variable(MILP, LatePerDayPos[1:num_tasks] >= 0); 
@variable(MILP, LatePerDayNeg[1:num_tasks] >= 0); 
    
@objective(MILP, Min, sum(DueDev[i] for i in 1:num_tasks)); 

@constraint(MILP, Earliness[i=1:num_tasks], DueDev[i]>= sum((Dates.value(d_task[i]-d_all[t])*X[t,i]) for t in 1:T_plan) );
@constraint(MILP, Tardiness[i=1:num_tasks], DueDev[i]>= sum((Dates.value(d_all[t]-d_task[i])*X[t,i]) for t in 1:T_plan) );
@constraint(MILP, CapacityRestriction[t=1:T_plan], sum( X[t,i] for i=1:num_tasks) <= cap );
@constraint(MILP, TaskExactlyOnce[i=1:num_tasks], sum(X[t,i] for t=1:T_plan) == 1 );
@constraint(MILP, OnlyOneHard[t=1:T_plan], sum( X[t,i]*task_types[i,1] for i in 1:num_tasks) <= 1);

@time optimize!(MILP);
```

```python
println("Objective value: ", JuMP.objective_value(MILP), " found after ", round(JuMP.solve_time(MILP), digits=2), " seconds. Relative gap is: ", JuMP.relative_gap(MILP))
```
